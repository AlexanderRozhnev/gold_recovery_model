```python
# Изучение общей информации таблиц будем делать по сценарию
def show_info(table):
    
    # общая информация
    table.info()
    
    # проверка наименований столбцов
    print('\nНаименования столбцов:')
    print(table.columns)
    
    # пропущенные значения
    print('\nПропущенные значения:')
    total_entries = table.shape[0]
    na_columns = table.isnull().sum().reset_index().rename(columns = {'index': 'column', 0: 'nan_qty'}, inplace = False)
    na_columns['na_percent'] = round((na_columns['nan_qty'] / total_entries * 100), 1)
    with pd.option_context('display.max_rows', None):
        display(na_columns)
    
    print('\nНаиболее значительные Пропущенные значения:')
    display(na_columns.loc[na_columns['na_percent'] > 1])
    
    # полные дубликаты
    print('\nПолные дубликаты:', table.duplicated().sum())
    
    # голова таблицы
    print('\nНачало таблицы данных:')
    display(table.head(5))
```

- Наименование признаков:

    - rougher — флотация

    - secondary_cleaner — вторичная очистка
    - final — финальные характеристики
- Возможные значения для блока [тип_параметра]:
    - input — параметры сырья
    - output — параметры продукта
    - state — параметры, характеризующие текущее состояние этапа
    - calculation — расчётные характеристики

    - air amount — объём воздуха

## Проверка правильности расчёта эффективности обогащения
Вычислите её на обучающей выборке для признака rougher.output.recovery.  
Найдите MAE между вашими расчётами и значением признака. Опишите выводы.  
MAE - Средняя абсолютная ошибка (Mean Absolute Error).  
C — доля золота в концентрате после флотации/очистки;  
F — доля золота в сырье/концентрате до флотации/очистки;  
T — доля золота в отвальных хвостах после флотации/очистки;    
Recovery - Эффективность обогащения.

```python
F = train_data['rougher.input.feed_au']
C = train_data['rougher.output.concentrate_au']
T = train_data['rougher.output.tail_au']

calculated_recovery = 100 * C * (F - T) / (F * (C - T))

error = mae(train_data['rougher.output.recovery'], calculated_recovery)
print('Средняя абсолютная ошибка между полученными данными\nи расчётным значением эффективности обогащения: {}'.format(error))
```

Средняя абсолютная ошибка между полученными данными и расчётным значением эффективности обогащения: 1.1498046842860452e-14.  
Т.е. значение ошибки мало, значения следует считать верными.

### Сравнение распределений размеров гранул сырья на обучающей и тестовой выборках.

Если распределения сильно отличаются друг от друга, оценка модели будет неправильной.  

Сравнение делаем на основе усредненных данных за сутки.

```python
# Размеры гранул сырья на обучающей и тестовой выборках до флотации
train_data['day'] = pd.to_datetime(train_data['date']).dt.floor('D')
test_data['day'] = pd.to_datetime(test_data['date']).dt.floor('D')
train_feed_size = train_data['rougher.input.feed_size'].groupby(train_data['day']).agg('mean')
test_feed_size =  test_data['rougher.input.feed_size'].groupby(test_data['day']).agg('mean')

sns.set_style("whitegrid")
f, axes = plt.subplots(1, 1, figsize=(14, 6))
sns.histplot(train_feed_size.loc[train_feed_size < 150], bins=30, kde= True, label = 'Обучающая', color='#4daf4a')
sns.histplot(test_feed_size.loc[test_feed_size < 150], bins=30, kde= True, label = 'Тестирующая', color='#ff7f00')
plt.title('Размеры гранул сырья на обучающей и тестовой выборках (до флотации)')
plt.xlabel('Размер гранул', fontsize=11)
plt.ylabel("Кол-во наблюдений", fontsize=11)
plt.legend(prop={'size': 11}, title = 'Выборки:')
sns.despine()
plt.show();
```

**Сравнение распределений**  

При расчете оценок выборки нас будет интересовать несмещенная оценка, поэтому параметр ddof=1

```python
# дисперсия считается функцией var
train_variance = np.var(train_feed_size, ddof=1)
test_variance = np.var(test_feed_size, ddof=1)
print('Дисперсия по обучающей выборке:', train_variance)
print('Дисперсия по тестовой выборке:', test_variance)
```

Дисперсии выборок отличаются незначительно.

*Нулевая гипотеза (равенство)*:  
Средний размер гранул в обоих выборках одинаков.  
*Альтернативная гипотеза (неравенство)*:  
Средний размер гранул в обоих выборках отличается.  

**Проверка гипотез с помощью двустороннего t-теста.**

```python
# Статистический уровень значимости
alpha = 0.05

# Проверка гипотезы
results = st.ttest_ind(train_feed_size, test_feed_size, equal_var = True)
# equal_var - считать ли равными дисперсии выборок, по умолчанию имеет значение True

print('p-значение: ', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

### Функция для вычисления итоговой sMAPE

```python
# Расчёт sMAPE
def smape(ground_truth, predictions):
    error = np.abs(ground_truth - predictions)
    scale = (np.abs(ground_truth) + np.abs(predictions)) / 2
    return np.mean(error / scale) * 100
```

```python
# Расчёт итоговой sMAPE
def final_smape(mape_rougher, mape_final):
    return 0.25 * mape_rougher + 0.75 * mape_final
```
